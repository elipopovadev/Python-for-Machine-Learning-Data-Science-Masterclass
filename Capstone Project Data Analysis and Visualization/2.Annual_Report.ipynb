## 7. ROIC & WACC

__ROIC__

```python
nopat = income_statement_df.loc["EBIT"] * (1 - income_statement_df.loc["Tax Provision"] / income_statement_df.loc["Pretax Income"])
nopat
```

```python
invested_capital = balance_sheet_df.loc['Long Term Debt And Capital Lease Obligation'] + balance_sheet_df.loc['Stockholders Equity']
invested_capital
```

```python
ROIC = nopat / invested_capital
ROIC
```

__WACC__

```python
def calculate_wacc(E, D, Ke, Kd, tax_rate):
    equity_weight = E / (E + D)
    debt_weight = D / (E + D)
    wacc = (equity_weight * Ke) + (debt_weight * Kd) * (1 - tax_rate)
    return wacc
```

```python
E = balance_sheet_df.loc['Stockholders Equity']
D = balance_sheet_df.loc['Long Term Debt']
Ke = 0.1
Kd = 0.06
tax_rate = 0.3
```

```python
wacc_result = calculate_wacc(E, D, Ke, Kd, tax_rate)
wacc_result
```
