## 4.1 balance_df_liabilities

```python
balance_df_liabilities = balance_sheet_df.loc[['Accounts Payable','Current Liabilities','Long Term Debt','Total Liabilities Net Minority Interest']]
balance_df_liabilities
```

```python
transposed_df_liabilities = balance_df_liabilities.T
transposed_df_liabilities
```

```python
transposed_df_liabilities.plot(kind = 'line',marker='o',subplots= True, color = 'r',figsize=(10,12), sharex= False, sharey=False)
plt.tight_layout()
plt.show()
```

```python
# Replace zero values with NaN to avoid division by zero
balance_df_liabilities_safe = transposed_df_liabilities.replace(0, pd.NA)

# Calculate percentage changes dynamically
balance_df_liabilities_percentage = balance_df_liabilities_safe.pct_change(axis=0, fill_method=None).mul(100)

# Display the result
balance_df_liabilities_percentage
```

```python
values = balance_df_liabilities_safe
percentages = balance_df_liabilities_percentage

# Create subplots
fig, axes = plt.subplots(nrows=values.shape[1], ncols=1,figsize=(10, 12), sharex=False, sharey=False)

for idx, liability in enumerate(values.columns):
    ax = axes[idx]
    ax.plot(values.index, values[liability], marker='o', color = 'r', label=f"{liability} Values")
    
    # Add annotations for percentage changes
    for year, value in enumerate(values[liability]):
        if not pd.isna(value):
            percentage = percentages[liability].iloc[year]
            if not pd.isna(percentage):
                ax.text(
                    values.index[year], value, f"{percentage:.2f}%", fontsize=10,
                    ha='center', va='bottom', color='black'
                )
    
    # Set labels and title for each subplot
    ax.set_title(f"{liability} Over Time", fontsize=14)
    ax.set_ylabel("Values (in Billions)", fontsize=12)
    ax.grid(True)
    ax.legend()

# Set shared x-axis label
axes[-1].set_xlabel("Years", fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

## 7. ROIC & WACC

__ROIC__

```python
nopat = income_statement_df.loc["EBIT"].replace({0: None}) * (1 - income_statement_df.loc["Tax Provision"] / income_statement_df.loc["Pretax Income"].replace({0: None}))
nopat
```

```python
invested_capital = balance_sheet_df.loc['Long Term Debt And Capital Lease Obligation'].replace({0: None}) + balance_sheet_df.loc['Stockholders Equity'].replace({0: None})
invested_capital
```

```python
ROIC = nopat / invested_capital
ROIC
```

__WACC__

```python
def calculate_wacc(E, D, Ke, Kd, tax_rate):
    equity_weight = E / (E + D)
    debt_weight = D / (E + D)
    wacc = (equity_weight * Ke) + (debt_weight * Kd) * (1 - tax_rate)
    return wacc
```

```python
E = balance_sheet_df.loc['Stockholders Equity'].replace({0: None})
D = balance_sheet_df.loc['Long Term Debt'].replace({0: None})
Ke = 0.1
Kd = 0.06
tax_rate = 0.3
```

```python
wacc_result = calculate_wacc(E, D, Ke, Kd, tax_rate)
wacc_result
```
